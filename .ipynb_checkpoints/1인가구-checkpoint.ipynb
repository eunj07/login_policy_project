```python
#   
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#    ( )
plt.rcParams['font.family'] = 'Malgun Gothic'

#  
df_seoul = pd.read_csv('__.csv')
df_gwanak = pd.read_excel(' 1.xlsx')

# --- [1]     1    ---

#  
seoul_one = df_seoul[df_seoul[''] == '']['1'].values[0]
gwanak_one = df_seoul[df_seoul[''] == '']['1'].values[0]

#  
ratio = gwanak_one / seoul_one * 100

# 
plt.figure(figsize=(6, 5))
plt.bar(['', ''], [seoul_one, gwanak_one], color=['lightgrey', 'skyblue'])
plt.title('    1 ')
plt.ylabel('1 ')
plt.text(1, gwanak_one + 5000, f'{ratio:.2f}%', ha='center', fontsize=12, color='blue')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


# --- [2]   1   ---

#   
df_gwanak_sorted = df_gwanak.sort_values(by='', ascending=False)

# 
plt.figure(figsize=(12, 6))
plt.bar(df_gwanak_sorted[''], df_gwanak_sorted[''], color='orange')
plt.title('  1 ')
plt.xticks(rotation=45)
plt.ylabel('1 ')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#    (Windows )
plt.rcParams['font.family'] = 'Malgun Gothic'

#  
df_seoul = pd.read_csv('__.csv')
df_gwanak = pd.read_excel(' 1.xlsx')

#    1 
seoul_total_households = df_seoul[df_seoul[''] == ''][''].values[0]
seoul_one_households = df_seoul[df_seoul[''] == '']['1'].values[0]
seoul_ratio = seoul_one_households / seoul_total_households * 100

#    1 
gwanak_total_households = df_seoul[df_seoul[''] == ''][''].values[0]
gwanak_one_households = df_seoul[df_seoul[''] == '']['1'].values[0]
gwanak_ratio = gwanak_one_households / gwanak_total_households * 100

# 
plt.figure(figsize=(6, 5))
plt.bar(['', ''], [seoul_ratio, gwanak_ratio], color=['gray', 'skyblue'])
plt.title('   1  ')
plt.ylabel('1  (%)')
plt.ylim(0, max(seoul_ratio, gwanak_ratio) + 5)
plt.text(0, seoul_ratio + 1, f'{seoul_ratio:.2f}%', ha='center', color='black')
plt.text(1, gwanak_ratio + 1, f'{gwanak_ratio:.2f}%', ha='center', color='blue')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
