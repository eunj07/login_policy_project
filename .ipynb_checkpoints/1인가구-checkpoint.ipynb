```python
# 필요한 라이브러리 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 한글 깨짐 방지 (윈도우 기준)
plt.rcParams['font.family'] = 'Malgun Gothic'

# 파일 불러오기
df_seoul = pd.read_csv('서울전체_관악구_가구비교.csv')
df_gwanak = pd.read_excel('관악구 1인가구.xlsx')

# --- [1] 서울 전체 대비 관악구의 1인가구 수 비율 시각화 ---

# 데이터 추출
seoul_one = df_seoul[df_seoul['구분'] == '서울시']['1인가구수'].values[0]
gwanak_one = df_seoul[df_seoul['구분'] == '관악구']['1인가구수'].values[0]

# 비율 계산
ratio = gwanak_one / seoul_one * 100

# 시각화
plt.figure(figsize=(6, 5))
plt.bar(['서울시', '관악구'], [seoul_one, gwanak_one], color=['lightgrey', 'skyblue'])
plt.title('서울 전체 대비 관악구 1인가구 수')
plt.ylabel('1인가구 수')
plt.text(1, gwanak_one + 5000, f'{ratio:.2f}%', ha='center', fontsize=12, color='blue')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


# --- [2] 관악구 동별 1인가구 수 시각화 ---

# 동 기준으로 정렬
df_gwanak_sorted = df_gwanak.sort_values(by='가구수', ascending=False)

# 시각화
plt.figure(figsize=(12, 6))
plt.bar(df_gwanak_sorted['동'], df_gwanak_sorted['가구수'], color='orange')
plt.title('관악구 동별 1인가구 수')
plt.xticks(rotation=45)
plt.ylabel('1인가구 수')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 한글 폰트 설정 (Windows 기준)
plt.rcParams['font.family'] = 'Malgun Gothic'

# 데이터 불러오기
df_seoul = pd.read_csv('서울전체_관악구_가구비교.csv')
df_gwanak = pd.read_excel('관악구 1인가구.xlsx')

# 서울시 전체 대비 1인가구 비율
seoul_total_households = df_seoul[df_seoul['구분'] == '서울시']['전체가구수'].values[0]
seoul_one_households = df_seoul[df_seoul['구분'] == '서울시']['1인가구수'].values[0]
seoul_ratio = seoul_one_households / seoul_total_households * 100

# 관악구 전체 대비 1인가구 비율
gwanak_total_households = df_seoul[df_seoul['구분'] == '관악구']['전체가구수'].values[0]
gwanak_one_households = df_seoul[df_seoul['구분'] == '관악구']['1인가구수'].values[0]
gwanak_ratio = gwanak_one_households / gwanak_total_households * 100

# 시각화
plt.figure(figsize=(6, 5))
plt.bar(['서울시', '관악구'], [seoul_ratio, gwanak_ratio], color=['gray', 'skyblue'])
plt.title('전체 가구 중 1인가구 비율 비교')
plt.ylabel('1인가구 비율 (%)')
plt.ylim(0, max(seoul_ratio, gwanak_ratio) + 5)
plt.text(0, seoul_ratio + 1, f'{seoul_ratio:.2f}%', ha='center', color='black')
plt.text(1, gwanak_ratio + 1, f'{gwanak_ratio:.2f}%', ha='center', color='blue')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
